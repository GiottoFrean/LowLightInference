### The Big Problem

In *B - Sample efficient maximisation*, I discussed how to sample a function in a low-dimensional space in a sample-efficient manner. A good answer is to choose points which maximise the expected improvement over your current best sample. You can do this in a straightforward way with a Gaussian Process (GP).

But what if your problem is instead the total distance you need to travel between samples? Imagine an autonomous robot which needs to move to each position iteratively. Then this method isn't close to optimal.
